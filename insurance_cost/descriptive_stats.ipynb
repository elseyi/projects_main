Health Insurance

Dataset is from kaggle and contains personal attributes and cost of medical insurance. This set can be used to see how insurance costs are affected by various factors.

This file is focused on descriptive stats

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## read in csv 

df = pd.read_csv("insurance.csv")
```

```python
print(df[['age','bmi','children','charges']].describe ())
```

```python
print(pd.crosstab(df['sex'], columns='count',normalize=True))
```

```python
print(pd.crosstab(df['smoker'], columns='count',normalize=True))
```

```python
print(pd.crosstab(df['region'], columns='count',normalize=True))
```

```python
freq_table1 = pd.crosstab(df['sex'], df['smoker'], normalize='index')
print(freq_table1)
```

```python
freq_table2 = pd.crosstab(df['sex'], df['region'], normalize='index')
print(freq_table2)
```

```python
age_charges = df.groupby('age')['charges'].mean()

# Plot the average charges by age
age_charges.plot(kind="line", figsize=(10, 6))

# Add labels and title for better clarity
plt.title("Average Charges by Age")
plt.xlabel("Age")
plt.ylabel("Average Charges")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
avg_charges_region = df.groupby('region')['charges'].mean()

# Plot the average charges by region
avg_charges_region.plot(kind="bar", figsize=(10, 6))

# Add labels and title for better clarity
plt.title("Average Charges by Region")
plt.xlabel("Region")
plt.ylabel("Average Charges")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
avg_charges = df.groupby('sex')['charges'].mean()

# Plot the average charges by region
avg_charges.plot(kind="bar", figsize=(10, 6))

# Add labels and title for better clarity
plt.title("Average Charges by Sex")
plt.xlabel("Sex")
plt.ylabel("Average Charges")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
avg_children_sex = df.groupby('sex')['children'].mean()

# Plot the average children by sex
avg_children_sex.plot(kind="bar", figsize=(10, 6))

# Add labels and title for better clarity
plt.title("Average Children by Sex")
plt.xlabel("Sex")
plt.ylabel("Average Children")
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.show()
```
